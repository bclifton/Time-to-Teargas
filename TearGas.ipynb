```python
for i, row in df1.iterrows():
    sections = row.section.split('; ')
    article_countries = set()
    article_regions = set()
    
    if 'commentisfree' in sections or 'Opinion'in sections or 'Books' in sections:
        continue
    elif 'uk-news' in sections or 'edinburgh' in sections:
        df1.loc[i,'country_1'] = 'united kingdom'
        continue
    elif 'us-news' in sections or 'U.S.' in sections or 'N.Y. / Region' in sections or 'N.Y./Region' in sections or 'New York and Region'in sections:
        df1.loc[i,'country_1'] = 'united states'
        continue
    elif 'australia-news' in sections:
        df1.loc[i,'country_1'] = 'australia'
        continue    
    # end known variables
    
#     elif 'World' in sections or 'world' in sections:
    elif 'World' in sections:

        # the Guardian's 'world' section includes the United States.
        
        if not pd.isnull(row.keywords):
            keywords = set(re.split('; |, ', row.keywords.lower()));
            kws = list(keywords)
            for kw in kws:
                if kw in countries:
                    article_countries.add(kw)
        url_strings = p.findall(row.web_url.lower(), re.IGNORECASE)
        for word in url_strings:
            if word in countries:
                article_countries.add(word)
            if word in regions:
                article_regions.add(word)  
    else:
        if not pd.isnull(row.keywords):
            keywords = set(re.split('; |, ', row.keywords.lower()));
            kws = list(keywords)
            for kw in kws:
                if kw in countries:
                    article_countries.add(kw)
        if not pd.isnull(row.lead_paragraph):
            words = set(p.findall(row.lead_paragraph.lower(), re.IGNORECASE))
            wrds = list(words)
            for wrd in wrds:
                if wrd in countries:
                    article_countries.add(wrd)
        
        test_words = []
        
        url_strings = p.findall(row.web_url.lower())
        headline_strings = p.findall(row.headline.lower())
        
#         print headline_strings
        
        for word in url_strings:
            test_words.append(word)
        for word in headline_strings:
            if word in peoples:
#                 print peoples[word]
                test_words.append(peoples[word])
        for word in lead_paragraph_strings:
            test_words.append(word)
#         test_words.append(url_strings)
#         test_words.append(headline_strings)
#         test_words.append(lead_paragraph_strings)
        
#         print set(test_words)
        

            
        
        for word in test_words:
            if word in alt_country_names:
                article_countries.add(alt_country_names[word])
            if word in countries:
                article_countries.add(word)
            if word in regions:
                article_regions.add(word)
    
#     print list(article_countries)  
    
    if 'palestinians' in article_countries or 'palestinian' in article_countries or 'west bank' in article_countries or 'gaza' in article_countries:
        if 'israel' in article_countries:
            article_countries = ['israel/palestine']
        else:
            article_countries = ['palestine']
            
    for index, country in enumerate(article_countries):
            col = 'country_' + str(index + 1)
            if country in alt_country_names:
                df1.loc[i, col] = alt_country_names[country]
            else:
                df1.loc[i, col] = country
  
            
```
